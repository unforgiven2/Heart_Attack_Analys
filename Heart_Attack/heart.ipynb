```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df=pd.read_csv("Heart.csv")
```

```python
sns.boxplot(y="age", data=df , color="salmon",saturation=1 )
```

```python
sns.pairplot(df)
```

```python
c=df.corr()
plt.figure(figsize=(20,10))
sns.heatmap(c, annot=True,fmt=".1f", robust=True, cmap="crest" )
```

```python
figure , ax = plt.subplots(figsize=(10,6))
df["age"].hist(color="#FAAE7B",bins=40)
plt.xlabel("Age")
plt.title("Age Dist* ");
```

```python
figure, ax = plt.subplots(figsize=(10,6))
df["oldpeak"].hist(color="#bcbddc",bins=20)
plt.xlabel("oldpeak")
plt.title("oldpeak Dist* ");
```

```python
figure, ax = plt.subplots(figsize=(10,6))
df["thalachh"].hist(color= "#fee0d2",bins=40)
plt.xlabel("thalachh")
plt.title("thalachh Dist* ");
```

```python
figure, ax = plt.subplots(figsize=(10,6))
df["chol"].hist(color="#fc9272",histtype="stepfilled",bins=40)
plt.xlabel("chol")
plt.title("chol Dist* ");
```

```python
plt.figure(figsize=(20,10))
sns.countplot(x=df["age"],saturation=1, palette=["#bcbddc", "#efedf5", "#fc9272", "#fee0d2",'#432371',"#FAAE7B"])
plt.title("COUNT OF PATIENTS AGE",fontsize=20)
plt.xlabel("AGE",fontsize=20)
plt.ylabel("COUNT",fontsize=20)
```

```python
plt.figure(figsize=(20,10))
sns.lineplot(y="trtbps",x="age",data=df, color='#432371')
plt.title("BLOOD PRESSURE WITH AGE",fontsize=20)
plt.xlabel("AGE",fontsize=20)
plt.ylabel("BLOOD PRESSURE",fontsize=20)
plt.show()
```

```python
plt.figure(figsize=(20,10))
sns.lineplot(y="chol",x="age",data=df,color="#fc9272")
plt.title("CHOLESTROL LEVEL  WITH AGE",fontsize=20)
plt.xlabel("AGE",fontsize=20)
plt.ylabel("CHOLESTROL LEVEL",fontsize=20)
plt.show()
```
